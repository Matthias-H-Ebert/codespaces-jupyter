# Lineare Regression - Proof of Concept

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Hier beliebige x und y Werte eingeben (es müssen nur gleich viele sein)

```python
x = np.array([ 105,  120,  132,  108])
y = np.array([1152, 1402, 1579, 1291])
print(x,y)
```

## Das Numpy-Array mit den x Werten muss zweidimensional werden
... da die scikit-learn-Klasse LineareRegression univariate y=f(x) und multivariate y=f(x1,x2,..,xn) Regression gleich behandet

```python
x = x.reshape(-1, 1)
print(x,y)
```

## Jetzt gehts los: 
Gesucht ist die Funktion y=mx+b, die für die gegebenen x und y Werte möglichst gut passt

```python
erg = LinearRegression().fit(x,y)
```

# Y-Achsenstartpunkt bei x=0 
b bzw. theta0 

```python
t0 = erg.intercept_
t0
```

## Steigung der Geraden
m bzw. theta1

```python
t1 = erg.coef_[0]
t1
```

# Zeichnen der Regressionsgerade

```python
plt.xlabel("Input Data")
plt.ylabel("Output Data")
plt.scatter(x,y,color="blue")
xcoords = np.linspace(min(x)-10, max(x)+10, 100)
ycoords = t0 + t1*xcoords
plt.title(f" y = {t0} + {t1}*x")
plt.plot(xcoords,ycoords, color="red")
plt.show()
```
